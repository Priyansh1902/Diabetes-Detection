```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
# -----------------------------------------------------------
# ------------ML ALgos--------------------------------------
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

All ML Clssificatin Models

```python
KNN_Model = KNeighborsClassifier(n_neighbors=5)
SVM_Model = SVC(kernel='linear')
DecisionTree_Model = DecisionTreeClassifier()
RandomForst_Model = RandomForestClassifier()
```

Importing The Data

```python
Data = pd.read_csv("./Diabetes Classification.csv")
Data.head(5)
```

Splitng The Taget and Features Apart

```python
X = Data[['Age', 'Gender', 'BMI', 'Blood Pressure', 'FBS', 'HbA1c', 'Family History of Diabetes', 'Smoking', 'Diet', 'Exercise']]
Y = Data['Diagnosis']
```

Checking For Null Data in Both 

```python
X.isnull().any()
```

```python
Y.isnull().any()
```

Converting The Data In Intergers For ML Model With Label Incoder

```python
# Label encode categorical features
label_encoder = LabelEncoder()

for column in X.columns:
    if X[column].dtype == 'object':
        X[column] = label_encoder.fit_transform(X[column])
```

```python
X
```

Spliting The Data For Training and Testing

```python
# Split the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

Training All The ML Models

```python
SVM_Model.fit(X_train,Y_train)

KNN_Model.fit(X_train,Y_train)

DecisionTree_Model.fit(X_train,Y_train)

RandomForst_Model.fit(X_train,Y_train)
```

Checking The Score Of Models With Accuracy Score

```python
#KNN
y_pred = KNN_Model.predict(X_test)

# Evaluate the accuracy
accuracy = accuracy_score(Y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
#SVM
y_pred = SVM_Model.predict(X_test)

# Evaluate the accuracy
accuracy = accuracy_score(Y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
#Decision Tree
y_pred = DecisionTree_Model.predict(X_test)

# Evaluate the accuracy
accuracy = accuracy_score(Y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
#Random Forest 
y_pred = RandomForst_Model.predict(X_test)

# Evaluate the accuracy
accuracy = accuracy_score(Y_test, y_pred)
print("Accuracy:", accuracy)
```
